# [SIR Disease Model](https://www.maa.org/press/periodicals/loci/joma/the-sir-model-for-spread-of-disease-the-differential-equation-model)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def dFdt(F, t, b, k):
    s, i, r = F
    return [
        -b*s*i,
        b*s*i-k*i,
        k*i
    ]
```

```python
b = 1/2
k = 1/3
s0 = 1
i0 = 1.27*(10**(-6))
r0 = 0
t = np.linspace(0, 140, 1000)

sol = odeint(func=dFdt, y0=[s0, i0, r0], t=t, args=(b, k))
```

```python
sSol, iSol, rSol = sol.T
```

```python
plt.plot(t, sSol, color='b')
plt.plot(t, iSol, color='r')
plt.plot(t, rSol, color='g')
plt.legend(labels=['S', 'I', 'R'])
plt.xlabel('Days')
plt.grid()
```
