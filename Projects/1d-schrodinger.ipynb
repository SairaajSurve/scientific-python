# [Numerically Finding the Eigenstates/Energies of a 1D QuantumÂ System](https://lukepolsonphysicsblog.wordpress.com/2020/10/29/example-post-3/)

$$mL^2V(y) = 1000(y-1/2)^2$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
```

```python
N = 2000
dy = 1/N
y = np.linspace(0, 1, N+1)
```

```python
def mL2V(y):
    return 1000* (y-1/2)**2
```

```python
V = mL2V(y)
plt.plot(y, V)
```

```python
def soln(dy, y):
    main_diag = 1/dy**2 + mL2V(y)[1:-1]
    off_diag = np.ones(len(main_diag)-1)-1/(2*(dy**2))
    w, v = eigh_tridiagonal(main_diag, off_diag)
    return w, v
```

```python
mL2E, psi = soln(dy, y)
psi = psi.T
```

```python
plt.figure(figsize=(15, 10))
plt.plot(y[1:-1], psi[0], color='r')
plt.plot(y[1:-1], psi[1], color='b')
plt.plot(y[1:-1], psi[2], color='g')
plt.plot(y[1:-1], psi[3], color='y')
plt.legend(labels=[f'$\psi_0$: $E_0$={mL2E[0]:.2f}', f'$\psi_1$: $E_1$={mL2E[1]:.2f}', f'$\psi_2$: $E_2$={mL2E[2]:.2f}', f'$\psi_3$: $E_3$={mL2E[3]:.2f}'], prop={'size': 12})
plt.grid()
```

```python
plt.figure(figsize=(15, 10))
plt.title('Probability Distributions')
plt.plot(y[1:-1], psi[0]**2, color='r')
plt.plot(y[1:-1], psi[1]**2, color='b')
plt.plot(y[1:-1], psi[2]**2, color='g')
plt.plot(y[1:-1], psi[3]**2, color='y')
plt.grid()
```
